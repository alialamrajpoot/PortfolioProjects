I have two datasets: Balance_Sheet.xlsx and Income_Statement.xlsx. Both these datasets have three columns in common:

**"Company"**: The company's ticker name.

**"comp_type"** The type of industry the company in question belongs to. It is either "tech" for companies in the technology industry, "fmcg" for companies in the fast-moving consumer goods industry, and "real_est" for companies in the real estate industry.

**"Year"**: The year the company's information is from.
The rest of the columns in the datasets contain information from the financial statement of the "Company" in question. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_excel("Balance_Sheet.xlsx")
df2 = pd.read_excel("Income_Statement.xlsx")
```

```python
df1.head()
```

```python
df2.head()
```

```python
merged_df = pd.merge(df1, df2, on=["company", "comp_type", "Year"])
```

```python
merged_df["comp_type"].values
```

    Will proceed to computing Equity Multiplier Ratio:

```python
merged_df["leverage_ratio"] = merged_df["Total Assets"]/merged_df["Total Stockholder Equity"]
```

```python
pd_ratios = merged_df[["leverage_ratio"]]
```

Proceeding to compute Operating Margin Ratio:

```python
merged_df["profitability_ratio"] = merged_df["Operating Income"]/merged_df["Total Revenue"]
```

```python
df_ratios = merged_df[["leverage_ratio", "profitability_ratio"]]
```

```python
df_ratios
```

```python
lowest_profitability = str(merged_df.pivot_table(index="comp_type", values = "profitability_ratio").idxmin()[0])
print(lowest_profitability)
```

```python
highest_leverage = str(merged_df.pivot_table(index="comp_type", values = "leverage_ratio").idxmax()[0])
print(highest_leverage)
```

```python
real_estate_df_ratios = merged_df[merged_df["comp_type"] == "real_est"][["leverage_ratio", "profitability_ratio"]]
```

```python
sns.regplot(data=real_estate_df_ratios, x='leverage_ratio', y='profitability_ratio', ci=None, line_kws={"color": "red"})
```

### This shows an upward positive correlation between profitability and leverage ratios
